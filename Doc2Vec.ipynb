```python
file = "data/german_news_articles_original_train_and_test_tokenized.parq"
data_all = load_dataframe(file)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

data_all.drop(columns=['label_encoded', 'label_onehot'], errors='ignore', inplace=True)
data_all['label_encoded'] = encoder.fit_transform(data_all['label'])

data_all.head(5)
```

```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
data_train_orig = data_all.loc[(data_all["split"] == "train")]
data_test_orig = data_all.loc[(data_all["split"] == "test")]
data_train_orig.head(5)
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
documents=[]

for index, row in data_train_orig.iterrows():
    documents.append(TaggedDocument(row['token_lemma'], f"{index}"))
    
documents[:1]
```

```python
model = Doc2Vec(vector_size=200, window=200, min_count=10, workers=10)
```

```python
model.build_vocab(documents)
model
```

```python
model.train(documents, total_examples=model.corpus_count, epochs=20)
```

```python
for row in range(10):
    text = data_test_orig.iloc[row, 5]
    label = data_test_orig.iloc[row, 8]
    
    inferred_vector = model.infer_vector(text)
    sims = model.dv.most_similar([inferred_vector], topn=len(model.dv))
    sims_doc_label = data_test_orig.iloc[int(sims[0][0]), 8]
    print(label, sims_doc_label, text[:5])
```

```python
y_test_pred = []
y_test = []
for index, row in data_test_orig.iterrows():
    text = row['token_lemma']
    label = row['label_encoded']
    inferred_vector = model.infer_vector(text)
    sims = model.dv.most_similar([inferred_vector], topn=len(model.dv))
    sims_doc_label = data_test_orig.iloc[int(sims[0][0]), 8]
    y_test_pred.append(sims_doc_label)
    y_test.append(label)
```

```python
len([p for p in zip(y_test_pred, y_test) if p[0] != p[1]])
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_test_pred)
print("Test:")
print(report)
```
